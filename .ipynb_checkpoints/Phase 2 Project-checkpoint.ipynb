### 4. Normalizing the data
It is important that our data is normally distributed as this will help in our final model. We will check for the normal distribution among the predictor variables and then normalize or scale the continuous variables as needed. 

```python
#Using a histogram to check for normal distribution
df_corre.hist(figsize=(18,15), bins='auto');
```

The predictor variables are quite skewed.  We can try to make them appear more normal using the log transformations.
We can deduce the following from the distribution of the data set.
1. We can also see that most houses condition are at 3.0.
2. Most houses have a low rating on the view.
3. Most of the houses that were sold had 1 and 2 floors.
4. Most of the houses that were sold had 1 and 2.5 bathrooms.
5. There's been an increase in the number of built over the years, with most of them been built around 2000.


```python
#Normalize our continuous variables using np.log
df_log = pd.DataFrame([])
df_log['price'] = np.log(df_corre['price'])
df_log['bedrooms'] = np.log(df_corre['bedrooms'])
df_log['bathrooms'] = np.log(df_corre['bathrooms'])
df_log['sqft_lot'] = np.log(df_corre['sqft_lot'])
df_log['floors'] = np.log(df_corre['floors'])
df_log.hist(figsize=[18,15], bins= 'auto');
```

```python
df_corre.info()
```

Now that we have normalized our data we can scale it. Some variables have a scale of up to 14 and the smallest 1.2, to ensure that their magnitudes are not too far a apart, we can use standardization. This will eliminate the chances of one variable looking more important than the others.

```python
# Create a function scaled that will take in the log tansformed data and scale it using standardization.
def scaled(logdata):
    for feature in logdata:
        scaled_feature = (logdata[feature] - np.mean(logdata[feature])) / np.sqrt(np.var(logdata[feature]))
        print (scaled_feature)
    
    return (scaled_feature)
```

```python
#Call the function to view the scaled data
scaled(df_log)
```

```python
# Create a new dataframe from the scaled data.
df_scaled = pd.DataFrame([])

for feature in df_log:
    df_scaled[feature] = scaled(df_log)
    print (df_scaled)
    
```

```python
df_scaled
```

Now to plot the scaled features on a histogram.


```python
df_scaled.hist(figsize=[18,15])
```

Our data is now scaled to a range between -1.0 and 2.5

### 5. Converting Categorical Features into Numeric Format
All our columns are numeric but they can still be categorical for example 'yr_built' is categorical even without confirming. From our previous step we can tell that 'view' and 'waterfront' are also categorical from the number of values, they only have 5 and 2 respectively. We can further confirm this using scatter plots.

```python
fig, axes = plt.subplots(nrows=1, ncols=4, figsize=(16,3))

for xcol, ax in zip(['view', 'condition', 'waterfront', 'yr_built'], axes):
    df.plot(kind='scatter', x=xcol, y='price', ax=ax, alpha=0.4, color='b')
```

The scatter plots confirm that these columns contain categorical columns since they form vertical lines opposed to numeric columns that would form clouds. We can proceed to transform them to a numeric format by generating dummies.

We will start by changing the float datatypes into integer type for purposes of labelling the columns. Generating dummies with float datatype generate columns separated with a '.'which will result into a syntax error in our moddel.

```python
df_corre['waterfront'] = df['waterfront'].apply(lambda x: int(x))
```

```python
df_corre['view'] = df['view'].apply(lambda x: int(x))
```

```python
waterfront_dummies = pd.get_dummies(df_corre['waterfront'], prefix='wtrfront', drop_first=True)
view_dummies = pd.get_dummies(df_corre['view'], prefix='view', drop_first=True)
condition_dummies = pd.get_dummies(df_corre['condition'], prefix='cndtn', drop_first=True)
yr_built_dummies = pd.get_dummies(df_corre['yr_built'], prefix='yr', drop_first=True)
```

```python
#Dropping original columns
df_dummies = df_corre.copy()
df_dummies = df_dummies.drop(['waterfront','view','condition','yr_built'], axis=1)
#Adding dummy columns
df_dummies= pd.concat([waterfront_dummies, view_dummies,condition_dummies,yr_built_dummies],axis=1)
df_dummies.head()
```

```python
df_dummies.dtypes
```

We will combine all our prepared data to form the final dataset that we will use for the modelling.

```python
#Combining the normalized continuous variables and the converted categorical variables
df_clean = pd.concat([df_scaled, df_dummies], axis=1)
df_clean
```

## Modeling

We can proceed to create a model using statsmodels with price as our dependent variable. In line with our objectives we will create a model showing the relationship between price, number of bedrooms and the view rating of the house.

We will proceed to interpret the model parameters and the significance of the variables on price. The waterfront is also a type of view therefore we can investigate its relationship with price as well. 

We have established a hypothesis: 
Alternative Hypothesis: The independent variables view and bedroom have a relationship with the price of the houses
Null Hypothesis: The independent variables view and bedroom do not have a relationship with increase the price of the houses
Our significance level is 0.05

We will check on the Assumptions for Linear Regression and correct them before proceeding to run our final model.

```python
#importing the libraries for modeling 
import statsmodels.api as sm
from statsmodels.formula.api import ols
```

```python
# Investigating the relationship between price and view, waterfront and bedroom combined.
formula = 'price ~ view_dummies+bedrooms+waterfront_dummies'
model = ols(formula=formula, data=df_clean).fit()
model.summary()
```

From the summary we can see that the p-value for most variables is below 0.05 stating that they are statistically significant except one view variable rating. This means that the combination of these variables affects the price of the houses. The r-squared value is at quite high at 1 indicating a high value for the goodness of fit for the model.

```python
# Investigating the relationship between price and the view.
formula = 'price ~ view_dummies'
model_1 = ols(formula=formula, data=df_clean).fit()
model_1.summary()
```

The p-value is below 0.05 for the view dummies except one view variable rating, however the other view ratings state that the view variable is statistically significant. We can reject the null hypothesis and state that the view affects the price of the houses.

```python
formula = 'price ~ bedrooms'
model_2 = ols(formula=formula, data=df_clean).fit()
model_2.summary()
```

Similar to the the view variable, the p-value is also below 0.05 meaning that the bedroom variable is statistically significant. We can reject the null hypothesis and state that the bedroom affects the price of the houses. The value of the intercept indicates a positive relationship between price and bedroom, the r-squared value is much higher than that of the view, the bedroom variable would be a better predicator than bedroom if we were to make predictions.
The positive coefficient values shows that there's a a positive relationship betwee the thwo such that when the number of bedrooms increase the price also increases.

```python
formula = 'price ~ waterfront_dummies'
model_3 = ols(formula=formula, data=df_clean).fit()
model_3.summary()
```

The p-value is less than 0.05 making it statistically signifcant just like the other variables and its r-squared value is much less than that of the bedrooms variable. This makes it our worst indicator so far. 

### Check for Normality

```python
import scipy.stats as stats
fig = sm.graphics.qqplot(model_1.resid, dist=stats.norm, line='45', fit=True)
```

Its clear that there's a violation of the normality assumption using the view variable since a good number of the points are outside the line the outliers may also be contributing to this.

```python
fig = sm.graphics.qqplot(model_2.resid, dist=stats.norm, line='45', fit=True)
```

Similar to the view model the bedroom model also violates the normality assumption.
