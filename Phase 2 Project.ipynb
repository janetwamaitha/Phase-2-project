### 4. Check for Multicollinearity

We need to identify highly correlated variables and drop them. This because high multicollinearity can cause misinterpretation of the model and small changes within the variables can lead to huge fluctuations in the model. We will check the correlation  among the independent variables.

```python
#Create a variable that contains all variables except the target variable 'price'
df_corre = df.iloc[:,1:13]
df_corre.head()
```

We can use a correlation matrix to check the correlation between the features. Correlation values lie between -1 and 1, a perfect negative correlation and a perfect positive correlation, respectively.

```python
# Use .corr() method to create the correlation matrix
df_corre.corr()
```

Since a high correlation is considered to lie between 0.7 and 0.8 we can use 0.75 as our mark to chech the high correlations. If it returns 'True' that is considered a high correlation.

```python
abs(df_corre.corr()) > 0.75
```

The following pairs bathrooms and sqft_living, grade and sqft_living are highly correlated. Therefore we can start by removing the 'sqft_living' that appears in both and the 'grade' since its correlation value to 'sqft_living' is higher compared to 'bathrooms.'

We can confirm the same using a heatmap.

```python
import seaborn as sns
sns.heatmap(df_corre.corr(), center=0);
```

```python
df_corre = df_corre.drop(['sqft_living', 'grade'], axis=1)
```

### 5. Normalizing the data
It is important that our data is normally distributed as this will help in our final model. We will check for the normal distribution among the predictor variables and then normalize or scale them as needed. 

```python
#Using a histogram to check for normal distribution
df_corre.hist(figsize=(18,15), bins='auto');
```

The predictor variables are quite skewed.  We can try to make them appear more normal using the log transformations.
We can deduce the following from the distribution of the data set.
1. We can also see that most houses condition are at 3.0.
2. Most houses have a low rating on the view.
3. Most of the houses that were sold had 1 and 2 floors.
4. Most of the houses that were sold had 1 and 2.5 bathrooms.


```python
df_log = pd.DataFrame([])
df_log['logbedrooms'] = np.log(df_corre['bedrooms'])
df_log['logbathrooms'] = np.log(df_corre['bathrooms'])
df_log['logsqft_lot'] = np.log(df_corre['sqft_lot'])
df_log['logfloors'] = np.log(df_corre['floors'])
# df_log['logview'] = np.log(df_corre['view'])
# df_log['logwaterfront'] = np.log(df_corre['waterfront'])
# df_log['logcondition'] = np.log(df_corre['condition'])
# df_log['logyr_built'] = np.log(df_corre['yr_built'])
df_log.hist(figsize=[18,15], bins= 'auto');
```

Now that we have normalized our data we can scale it, we can use standardization.

```python
# Create a function scaled that will take in the log tansformed data and scale it using standardization.
def scaled(logdata):
    for feature in logdata:
        scaled_feature = (logdata[feature] - np.mean(logdata[feature])) / np.sqrt(np.var(logdata[feature]))
        print (scaled_feature)
    
    return (scaled_feature)
```
