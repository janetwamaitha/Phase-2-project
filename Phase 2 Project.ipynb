### 4. Normalizing the data
It is important that our data is normally distributed as this will help in our final model. We will check for the normal distribution among the predictor variables and then normalize or scale the continuous variables as needed. 

```python
#Using a histogram to check for normal distribution
df_corre.hist(figsize=(18,15), bins='auto');
```

The predictor variables are quite skewed.  We can try to make them appear more normal using the log transformations.
We can deduce the following from the distribution of the data set.
1. We can also see that most houses condition are at 3.0.
2. Most houses have a low rating on the view.
3. Most of the houses that were sold had 1 and 2 floors.
4. Most of the houses that were sold had 1 and 2.5 bathrooms.
5. There's been an increase in the number of built over the years, with most of them been built around 2000.


```python
#Normalize our continuous variables using np.log
df_log = pd.DataFrame([])
df_log['bedrooms'] = np.log(df_corre['bedrooms'])
df_log['bathrooms'] = np.log(df_corre['bathrooms'])
df_log['sqft_lot'] = np.log(df_corre['sqft_lot'])
df_log['floors'] = np.log(df_corre['floors'])
df_log.hist(figsize=[18,15], bins= 'auto');
```

```python
df_corre.info()
```

Now that we have normalized our data we can scale it. Some variables have a scale of up to 14 and the smallest 1.2, to ensure that their magnitudes are not too far a apart, we can use standardization. This will eliminate the chances of one variable looking more important than the others.

```python
# Create a function scaled that will take in the log tansformed data and scale it using standardization.
def scaled(logdata):
    for feature in logdata:
        scaled_feature = (logdata[feature] - np.mean(logdata[feature])) / np.sqrt(np.var(logdata[feature]))
        print (scaled_feature)
    
    return (scaled_feature)
```

```python
# for feature in df_log:
#     scaled_feature = (df_log[feature] - np.mean(df_log[feature])) / np.sqrt(np.var(df_log[feature]))
#     print (scaled_feature)
```

```python
#Call the function to view the scaled data
scaled(df_log)
```

```python
# Create a new dataframe from the scaled data.
df_scaled = pd.DataFrame([])

for feature in df_log:
    df_scaled[feature] = scaled(df_log)
    print (df_scaled)
    
```

### 5. Converting Categorical Features into Numeric Format
All our columns are numeric but they can still be categorical for example 'yr_built' is categorical even without confirming. From our previous step we can tell that 'view' and 'waterfront' are also categorical from the number of values, they only have 5 and 2 respectively. We can further confirm this using scatter plots.

```python
fig, axes = plt.subplots(nrows=1, ncols=4, figsize=(16,3))

for xcol, ax in zip(['view', 'condition', 'waterfront', 'yr_built'], axes):
    data.plot(kind='scatter', x=xcol, y='price', ax=ax, alpha=0.4, color='b')
```

The scatter plots confirm that these columns contain categorical columns since they form vertical lines opposed to numeric columns that would form clouds. We can proceed to transform them to a numeric format by generating dummies.

```python
waterfront_dummies = pd.get_dummies(df_corre['waterfront'], prefix='wtrfront', drop_first=True)
view_dummies = pd.get_dummies(df_corre['view'], prefix='view', drop_first=True)
condition_dummies = pd.get_dummies(df_corre['condition'], prefix='cndtn', drop_first=True)
yr_built_dummies = pd.get_dummies(df_corre['yr_built'], prefix='yr', drop_first=True)
```

```python
#Dropping original columns
df_dummies = df_corre.copy()
df_dummies = df_dummies.drop(['waterfront','view','condition','yr_built'], axis=1)
#Adding dummy columns
df_dummies= pd.concat([waterfront_dummies, view_dummies,condition_dummies,yr_built_dummies],axis=1)
df_dummies.head()
```

```python
df_dummies.dtypes
```
